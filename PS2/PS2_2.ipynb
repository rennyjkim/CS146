```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 1, 101)
Y = (X ** 6) * (1 - X) ** 4
plt.plot(X, Y)
xmax = X[np.argmax(Y)]
plt.axvline(x = xmax, color = 'b', label = 'max')
plt.legend()
plt.title('Likelhood for various theta, n=10')
plt.xlabel('theta')
plt.ylabel('Liklihood')
plt.show()
```
